```python
# Jupyter Notebook

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Define the data as a string
df = pd.read_csv("../results/results.csv")
# Load the data into a pandas DataFrame

# Convert 'Timestamp' to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Convert 'Time Taken' to a common unit (milliseconds)
df['Time Taken'] = df['Time Taken'].apply(lambda x: float(x.replace('ms','')) if 'ms' in x else float(x.replace('ns','')) / 1e6)

# Display the first few rows of the DataFrame
print(df.head())

# Plotting the data
plt.figure(figsize=(14, 7))

# Plot Time Taken vs Iteration for both directories
for dir_path in df['DirPath'].unique():
    dir_df = df[df['DirPath'] == dir_path]
    plt.plot(dir_df['Iteration'], dir_df['Time Taken'], marker='o', label=dir_path)

plt.xlabel('Iteration')
plt.ylabel('Time Taken (ms)')
plt.title('Time Taken for File Processing per Iteration')
plt.legend()
plt.grid(True)
plt.show()
```
